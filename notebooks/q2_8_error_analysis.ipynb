# Q2.8 Error Analysis

```python
import sys
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import wandb
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

project_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.utils.data_loader import load_dataset
from src.inference import load_model

ENTITY = 'anandhakrishnanm21-indian-institute-of-technology-madras'
```

```python
PROJECT = 'q2_8_error_analysis'
run = wandb.init(project=PROJECT, name='mnist_error_analysis', reinit=True)
model = load_model(str(project_root / 'src' / 'best_model.npy'))
_, _, _, _, X_test, Y_test = load_dataset('mnist')
logits = model.forward(X_test)
y_true = np.argmax(Y_test, axis=1)
y_pred = np.argmax(logits, axis=1)
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(7, 7))
ConfusionMatrixDisplay(cm).plot(ax=ax, cmap='Blues', colorbar=False)
wandb.log({'confusion_matrix': wandb.Image(fig)})
plt.close(fig)
wandb.finish()
```
