# Q2.1 Data Exploration
Log sample tables directly to W&B.

```python
import sys
from pathlib import Path
import numpy as np
import wandb

project_root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from src.utils.data_loader import load_dataset

ENTITY = 'anandhakrishnanm21-indian-institute-of-technology-madras'
```

```python
PROJECT = 'q2_1_data_exploration'
run = wandb.init(project=PROJECT, name='mnist_samples', reinit=True)
X_train, Y_train, *_ = load_dataset('mnist')
labels = np.argmax(Y_train, axis=1)

table = wandb.Table(columns=['class_id', 'sample_idx', 'image'])
for c in range(10):
    idxs = np.where(labels == c)[0][:5]
    for j, idx in enumerate(idxs):
        table.add_data(int(c), int(j + 1), wandb.Image(X_train[idx].reshape(28, 28)))

wandb.log({'mnist_class_samples': table})
wandb.finish()
print('Logged class samples to W&B.')
```
